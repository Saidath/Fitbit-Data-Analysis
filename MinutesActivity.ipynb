```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/minuteCaloriesNarrow_merged.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/minuteIntensitiesNarrow_merged.csv")
df3 = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/minuteMETsNarrow_merged.csv")
df4 = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/minuteStepsNarrow_merged.csv")
```

```python
merged_df = pd.merge(df1, df2, on=['Id', 'ActivityMinute'], how='outer')
merged_df= pd.merge(merged_df, df3, on=['Id', 'ActivityMinute'], how='outer')
df= pd.merge(merged_df, df4, on=['Id', 'ActivityMinute'], how='outer')
df.head()
```

```python
df['ActivityMinute'] = pd.to_datetime(df['ActivityMinute'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
df['Calories'] = df['Calories'].apply(lambda x: max(x, 0))
df['Intensity'] = df['Intensity'].apply(lambda x: max(x, 0))
df['METs'] = df['METs'].apply(lambda x: max(x, 0))
df['Steps'] = df['Steps'].apply(lambda x: max(x, 0))
```

```python
df.to_csv('/content/drive/MyDrive/Data Analytics Internship/Week 8/Cleaned/MinutesActivty.csv', index=False)
```
