```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/hourlyCalories_merged.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/hourlyIntensities_merged.csv")
df3 = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/hourlySteps_merged.csv")
```

```python
merged_df = pd.merge(df1,df2, on=["Id","ActivityHour"])
Hourly_Activity_merged_df = pd.merge(merged_df, df3, on=["Id","ActivityHour"])
```

```python
Hourly_Activity_merged_df['ActivityHour'] = pd.to_datetime(Hourly_Activity_merged_df['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
Hourly_Activity_merged_df['Calories'] = Hourly_Activity_merged_df['Calories'].apply(lambda x: max(x, 0))
Hourly_Activity_merged_df['TotalIntensity'] = Hourly_Activity_merged_df['TotalIntensity'].apply(lambda x: max(x, 0))
Hourly_Activity_merged_df['AverageIntensity'] = Hourly_Activity_merged_df['AverageIntensity'].apply(lambda x: max(x, 0))
Hourly_Activity_merged_df['StepTotal'] = Hourly_Activity_merged_df['StepTotal'].apply(lambda x: max(x, 0))
```
