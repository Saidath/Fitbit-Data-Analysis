```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data Analytics Internship/Week 8/Raw Data/data/dailyActivity_merged.csv")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'], format='%m/%d/%Y')
```

```python
df.head()
```

```python
df['TotalSteps'] = df['TotalSteps'].apply(lambda x: max(x, 0))
```

```python
df['TotalDistance'] = df['TotalDistance'].apply(lambda x: max(x, 0))
df['Calories'] = df['Calories'].apply(lambda x: max(x, 0))
```

```python
df.to_csv('cleaned_fitness_data.csv', index=False)
```
